```python
import pandas as pd
import re

# Load sample sheet
sample_sheet = pd.read_csv("../config/sample-sheet.csv", sep=";")

def getLane(sample):
    """Return sample's lane corresponding to given file."""
    return re.search("\.(\d)\.bam$", sample).group(1)

def getBamToQC(token):
    """Return a list containing all files expected as input for MultiQC command."""
    SAMPLES = sample_sheet['samples']
    out = []
    SAMPLE = sample_sheet.set_index("samples", drop = False)
    for sample in SAMPLE["samples"]:
        for bam in SAMPLE.at[sample, "somatic_path"].split(" "):
            template = "results/{token}/quality_control/{sample}_D.{lane}_fastqc.html".format(token=token, sample=sample, lane=getLane(bam))
            out.append(template)
        for bam in SAMPLE.at[sample, "germline_path"].split(" "):
            template = "results/{token}/quality_control/{sample}_G.{lane}_fastqc.html".format(token=token, sample=sample, lane=getLane(bam))
            out.append(template)
    return out

getBamToQC("token")
```
